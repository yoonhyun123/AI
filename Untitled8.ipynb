```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from google.colab import files

# 1. CSV 파일 업로드
uploaded = files.upload()

# 2. 업로드한 CSV 파일 로드
# 업로드한 파일의 이름은 uploaded.keys()로 확인 가능. 예를 들어 '이분로지스틱회귀.csv'라는 파일을 업로드했다면:
data = pd.read_csv(list(uploaded.keys())[0], encoding='cp949')

# 3. 독립 변수와 종속 변수 설정 (성별, 식전 혈당, 감마지티피, 음주 여부를 독립 변수로 사용)
X = data[['성별', '식전혈당(공복혈당)', '감마지티피', '음주여부']]
y = np.random.choice([0, 1], size=len(data))  # 임의로 종속 변수 설정 (당뇨 여부, 실제 데이터에서 수정 가능)

# 4. 상수항 추가 (회귀 분석에서 필요)
X = sm.add_constant(X)

# 5. 이분형 로지스틱 회귀 모델 학습
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# 6. 모델 요약 출력
print(result.summary())

# 7. P값, Exp(B), 가중치 출력
params = result.params
p_values = result.pvalues
odds_ratios = np.exp(params)

print("\n가중치 (Coefficients):\n", params)
print("\nP값 (P-values):\n", p_values)
print("\nExp(B) (Odds Ratios):\n", odds_ratios)
```

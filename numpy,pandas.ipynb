```python
import numpy as np

list=[1,2,3,4]
a=np.array(list)
print('a.shape:',a.shape)
print('a[0]',a[0])

b=np.array([[1,2,3],[4,5,6]])
print('b.shape:',b.shape)
print('b[0,0]',b[0,0])
print('b[0]:',b[1])
```

```python
a=np.zeros((2,3)) #0인 n차원
print('a:\n',a)

b=np.ones((2,3)) #1인 n차원
print('b:\n',b)

c=np.full((2,3),7) #값 n차원
print('c:\n',c)

d=np.eye(2)   #단위행렬
print('d:\n',d)

```

```python
a= np.array([1,2],dtype=np.float64)  #dtype 옵션을 추가하면 자료형 지정
print('a.dtype:',a.dtype)

b= a.astype(np.int8)
print(b.dtype)
```

```python
arr = np.array([[0, 1, 2], [3, 4, 5]])

print('type(arr):', type(arr))
print('arr.ndim:', arr.ndim)
print('arr.dtype:', arr.dtype)
print('arr.itemsize:', arr.itemsize)
print('arr.size:', arr.size)
print('arr.nbytes:', arr.nbytes)
print('arr.T:\n', arr.T)
print('arr.shape:', arr.shape)
```

```python
#1차원 배열 생성하기
a = np.arange(8)
print('a\n', a)

#다차원 배열로 변경하기
a.shape = (2,4)
print('shape\n', a)

#1차원 배열로 변경하기
print('flatten\n', a.flatten())

#resize 함수로 모양 변경하기
a.resize((4,2))
print('resize\n', a)
```

```python
a = np.array([[0, 1, 2], [3, 4, 5]])
print('a\n', a)

b = a.transpose()
print('b\n', b)

c = a.T  #transpose=T
print('c\n', c)
```

```python
import pandas as pd

a=pd.Series([1,2,3,4])
print(a)

b=pd.Series([1,2,3],index=['a','b','c'])
print(b)
```

```python
# 파일 복사하기
input_file = 'd:/phones.txt'  # 입력 파일 경로
output_file = 'd:/temp.txt'   # 출력 파일 경로

# 입력 파일 읽고 출력 파일에 쓰기
try:
    with open(input_file, 'r', encoding='utf-8') as infile:
        content = infile.read()

    with open(output_file, 'w', encoding='utf-8') as outfile:
        outfile.write(content)

    print(f"파일이 성공적으로 복사되었습니다: {output_file}")

except FileNotFoundError:
    print(f"입력 파일을 찾을 수 없습니다: {input_file}")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

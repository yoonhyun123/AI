```python
bream_len=[242, 290, 340, 363, 430, 450, 500, 390, 450, 500,
 475, 500, 500, 340, 600, 600, 700, 700, 610, 650,
 575, 685, 620, 680, 700, 725, 720, 714, 850, 1000,
 920, 955, 925, 975, 950]
bream_wgt=[23.2, 24, 23.9, 26.3, 26.5, 26.8, 26.8, 27.6, 27.6, 28.5,
 28.4, 28.7, 29.1, 29.5, 29.4, 29.4, 30.4, 30.4, 30.9, 31,
 31.3, 31.4, 31.5, 31.8, 31.9, 31.8, 32, 32.7, 32.8, 33.5,
 35, 35, 36.2, 37.4, 38]

import matplotlib.pyplot as plt

plt.scatter(bream_len,bream_wgt)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

```

```python
smelt_len=[6.7, 7.5, 7, 9.7, 9.8, 8.7, 10, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
smelt_wgt=[9.3, 10, 10.1, 10.4, 10.7, 10.8, 11.3, 11.3, 11.4, 11.5, 11.7, 12.1, 13.2, 13.8]

import matplotlib.pyplot as plt
plt.scatter(smelt_len,smelt_wgt)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length=bream_len+smelt_len
weight=bream_wgt+smelt_wgt

fish_data=[[1,t] for l,t in zip(length,weight)]
print(fish_data)
fish_target=[1]*35+[0]*14
print(fish_target)

from sklearn.neighbors import KNeighborsClassifier

kn=KNeighborsClassifier()
kn.fit(fish_data,fish_target)
kn.score(fish_data,fish_target)
kn.predict([[30,600]])

kn49=KNeighborsClassifier(n_neighbors=49)

kn49.fit(fish_data,fish_target)
kn49.score(fish_data,fish_target)
print(35/49)

```
